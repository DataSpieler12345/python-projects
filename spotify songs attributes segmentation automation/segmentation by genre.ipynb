```python
# import the libraries 
import pandas as pd
from pathlib import Path
```

```python
# define the rute of the spotify dataset 
file_path = Path('./data/Spotify_Song_Attributes.csv')
```

```python
# read the file with pandas
df = pd.read_csv(file_path)
```

```python
# obtain the info by unique genre
genre = df['genre'].unique()
```

```python
# for each genre, we create a sub-dataframe and save it in the corresponding folder
for genero in genre:
    print(genero)
```

```python
# Remove “nan” gender from the list
genre = [g for g in genre if isinstance(g, str) and g.strip().lower() != 'nan']

# Check the current length of the list
print(len(genre))
```

```python
import pandas as pd
from pathlib import Path
import re

# Suppose that 'df' is your main DataFrame and 'genre' is the list of genres.
for genero in genre:
    # Replace invalid characters in the gender name
    genero_cleaned = re.sub(r'[^a-zA-Z0-9_]', '_', genero)
    
    genero_df = df[df['genre'] == genero]
    
    # Create the folder of the genre if it does not exist
    genero_dir = Path(f'./allgenre/{genero_cleaned}')
    genero_dir.mkdir(parents=True, exist_ok=True)
    
    # Obtain unique records of the genus
    unique_records = genero_df.drop_duplicates()
    
    # Save the unique records in a CSV file in the genre folder
    csv_file_path = genero_dir / f'Spotify_Song_Attributes_{genero_cleaned}.csv'
    unique_records.to_csv(csv_file_path, index=False)

print("523 iteration are completed! Check it out!.")
```
